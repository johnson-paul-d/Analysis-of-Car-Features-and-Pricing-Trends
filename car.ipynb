```python
sns.pairplot(df)
```

#### Insights
1. Engine size correlates positively with MSRP and negatively with MPG. 
2. Higher MSRP vehicles are generally more popular but less fuel-efficient. 
3. Most vehicles have 4-6 engine cylinders, with higher MSRP outliers. 
4. Highway MPG tends to be higher than city MPG consistently. 
5. Recent years show increased popularity and varied engine sizes.

```python
H = df[['Engine Cylinders', 'Engine HP', 'city mpg', 'Popularity']]
corr = H.corr()
```

```python
sns.heatmap(corr,annot=True)
```

### Insights
Here are five insights from the correlation heatmap:

1. Engine Cylinders and Engine HP have a strong positive correlation (0.72).
2. City MPG has a negative correlation with Engine Cylinders (-0.51).
3. City MPG is inversely related to Engine HP with -0.32 correlation.
4. Popularity has weak correlations with other variables, close to zero.
5. More engine cylinders tend to reduce fuel efficiency (negative MPG correlation).

```python
df.columns
```

```python
sns.lineplot(df, x='Engine HP', y='MSRP')
```

#### Insights
**Line Plot of MSRP by Engine HP**:

- There is a strong positive relationship between MSRP and engine horsepower; as horsepower increases, so does MSRP.
- High horsepower vehicles (over 600 HP) have a much wider range in MSRP, with some extreme outliers reaching very high prices.
- The trend suggests that luxury or performance-focused vehicles with high horsepower command a premium in price.

```python
sns.lineplot(df, x='Number of Doors', y='MSRP')
```

#### insights

**Line Plot of MSRP by Number of Doors**:

- Vehicles with 2 doors generally have the highest MSRP, potentially due to sports and luxury models.
- The lowest MSRP is seen in vehicles with 3 doors, suggesting these may be less expensive, possibly smaller or economy models.
- Vehicles with 4 doors have higher MSRPs than 3-door vehicles, potentially reflecting the market demand for family cars.

```python
g = sns.FacetGrid(df, col='Vehicle Size'); g.map(plt.hist, 'MSRP')
```

#### Insights
 **Histogram of MSRP by Vehicle Size**:

- Most compact and midsize vehicles are concentrated in the lower MSRP range.
- Large vehicles tend to have slightly higher MSRPs, though they still cluster in the lower end of the MSRP scale.
- All three vehicle sizes have a right-skewed distribution, with few high-priced outliers.


```python
sns.scatterplot(data=df,x='MSRP',y='city mpg')
```

#### Insights
**Scatter Plot of City MPG vs. MSRP**:

- Lower MSRP vehicles generally have a higher city MPG.
- There is an inverse relationship between MSRP and city MPG, with high-priced vehicles tending to have lower fuel efficiency.
- Outliers are visible where some high-MSRP vehicles achieve relatively high MPG, but these are uncommon.

```python
sns.violinplot(df, x='Vehicle Size', y='city mpg')
```

#### Insights
**Violin Plot of City MPG by Vehicle Size**:

- Compact vehicles tend to have the highest city MPG range, with several outliers reaching high MPG values.
- Midsize vehicles show a more consistent and narrower range around the lower end of city MPG compared to compact vehicles.
- Large vehicles generally have lower city MPG, with a tight distribution around a lower average.
